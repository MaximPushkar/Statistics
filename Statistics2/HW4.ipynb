### Setup

```python
import numpy as np
import scipy as sci
import pandas as pd
import matplotlib.pyplot as plt

import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

### 1) 

Таблиця 3.1. Дослідіти модель лінійної регресії $y_1$ на $x_1$.

```python
names = ["n", "x1", "y1"]
df = pd.read_table("data/T3_1_HEIGHTWT.DAT", header=None, sep="\s+", names=names)
df
```

```python
x = df["x1"]
y = df["y1"]
```

```python
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
model.summary()
```

### 2)

Таблиця 3.4. Дослідити модель множинної регресії $y_1$ на $(x_1,x_2,x_3)$

```python
names = ["n", "y1", "y2", "x1", "x2", "x3"]
df = pd.read_table("data/T3_4_DIABETES.DAT", header=None, sep="\s+", names=names)
df.head(10)
```

```python
x = df[["x1", "x2", "x3"]]
y = df[["y1"]]
```

```python
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
model.summary()
```

### 3)

Таблиця 3.4. Дослідити модель багатовимірної множинної регресії $(y_1,y_2)$ на $(x_1,x_2,x_3)$.

```python
names = ["n", "y1", "y2", "x1", "x2", "x3"]
df = pd.read_table("data/T3_4_DIABETES.DAT", header=None, sep="\s+", names=names)
df.head(7)
```

```python
y = df[["y1", "y2"]]
x = df[["x1", "x2", "x3"]]
```

```python
model = LinearRegression() 
y_pred = model.fit(x, y).predict(x)

print("Regression results:")
print(f"Mean squared error = {metrics.mean_squared_error(y, y_pred):.3f}")
print(f"R_2 = {metrics.r2_score(y, y_pred):.3f}")
```
